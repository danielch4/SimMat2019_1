```python
#Ejercicio 1
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
sym.var('x')
```

```python
def g(x):
    return x**3
```

```python
dg=sym.diff(g(x),x)
dg
```

```python
xc=sym.solve(dg, x)
xc
```

```python
gnum=sym.lambdify(x,g(x))
xnum=np.linspace(-10,10,200)
ynum=gnum(xnum)
```

```python
plt.figure(figsize=(8,6))
plt.plot(xnum, g(xnum), 'k', label='$y=x^3$')
plt.plot([0],[0], '*y', ms=10, label='$(0,0)$')

plt.xlabel('$x$')
plt.ylabel('$f=x^3$')
plt.legend('f=x^3')
plt.grid()
```

```python
#Ejercico 2
def f(x):
    return x**3-3*x
```

```python
fsym=x**3-3*x
fsym
```

```python
df=sym.diff(fsym,x)
df
```

```python
xb=sym.solve(df,x)
xb
```
